####3 Funciones Normales y Funciones Lambda
1. a. Definir una función normal llamada multiplicar_por_dos que tome un número y lo multiplique por 2. Utilizar esta función para crear una nueva lista llamada dobles, que contenga los dobles de los números en la lista numeros.
2. b. Definir una función lambda que haga lo mismo y utilizarla para crear una nueva lista llamada dobles_lambda, que contenga los dobles de los números en la lista numeros.

```python
#2. a
print(notas)
notas["pepe"] = 1
print(notas)

#2. b
for numero in numeros:
  if not (numero % 2):
    print(numero)
#2. c
print(meses[0], meses[-1])

print("------")

#3. a
def multiplicar_por_dos (numero):
  return numero*2


dobles = []
for numero in numeros:
  dobles.append(multiplicar_por_dos(numero))
print("Normal:")
print(dobles)

#3. b
dobles_lambda = []
doble_lambda = lambda numero: numero*2
for numero in numeros:
  dobles_lambda.append(doble_lambda(numero))
print("En lambda: ")
print(dobles_lambda)
```
