####Actividad 1: Identificación de Valores Nulos y Duplicados

Objetivos:

Practicar la carga datos en un DataFrame utilizando Pandas.

Identificar y contar valores nulos en diferentes columnas del DataFrame.

Detectar y contar registros duplicados.

Generar un informe que resuma los hallazgos de la limpieza de datos.


####Consigna:

- 1. Descargar los datos en formato CSV y cargarlos en un DataFrame en Python usando Pandas. Utilizar el método isnull() para identificar las filas con datos faltantes y contar el número de valores nulos por columna.

- 2. Usar el método duplicated() para identificar las filas duplicadas y contar cuántas filas duplicadas hay en total.

- 3. Crear un informe que incluya:

    - La cantidad total de registros en el DataFrame.

     - La cantidad total de valores nulos por columna.

    - La cantidad de filas duplicadas.

- 4. Crear un dataframe de los registros duplicados para que se vea el contenido.

```python
#1:
import pandas as pd
#Descargar los datos en formato CSV y cargarlos en un DataFrame en Python usando Pandas.
df = pd.read_csv("/satis_clientes.csv")
#Utilizar el método isnull() para identificar las filas con datos faltantes y contar el número de valores nulos por columna.
print(df.isnull().sum())
print("----"*40)

#2:
#Usar el método duplicated() para identificar las filas duplicadas
print(df.duplicated())
#y contar cuántas filas duplicadas hay en total.
print(df.duplicated().sum())
print("----"*40)
'''
#3
Crear un informe que incluya:
La cantidad total de registros en el DataFrame.

La cantidad total de valores nulos por columna.

La cantidad de filas duplicadas.
'''
total_registros = (df.shape)
print(f"total registros : {total_registros}\n")

nulos_por_columna = df.isnull().sum()
print(f"nulos por columna: \n{nulos_por_columna}\n")

filas_duplicadas = df.duplicated().sum()
print(f"filas duplicadas: {filas_duplicadas}\n")
print("----"*40)


#4.Crear un dataframe de los registros duplicados para que se vea el contenido.

duplicadas = df[df.duplicated()]
print(duplicadas)
```
