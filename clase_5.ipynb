####Actividad 2: Corrección de Tipos de Datos y Normalización

Objetivos:

Aplicar los criterios y métodos apropiados para obtener tipos de datos consistentes y normalizados.


Consigna:

Cargar el conjunto de datos.

Corregir el tipo de datos de la columna de precio a un tipo numérico.

Normalizar el nombre del producto para que todas las entradas estén en minúsculas y sin caracteres especiales

```python
#1.Cargar el conjunto de datos.

sheet_id = "1KDJRNV5f1wpS42xU0ELTSKbyJ0fy5eRm8jhzfChUH5s"
#concatenar con la cadena completa
url = f"https://docs.google.com/spreadsheets/d/{sheet_id}/export?format=csv"

import pandas as pd

#el df obtiene dato del parametro url -> url obtiene de > sheed id
df = pd.read_csv(url)
df.head()#para probar si levanto OK

print(df)

print("----"*40)

#Corregir el tipo de datos de la columna de precio a un tipo numérico.
df['Edad'].dtype #consulto tipo de dato
df = df.dropna(subset=['Edad']) #borro los vacios

df['Edad'] = df['Edad'].astype(int)

print(df)
print("----"*40)
#Normalizar el nombre del producto para que todas las entradas estén en minúsculas y sin caracteres especiales
df['Nombre'] = (
    df['Nombre']
        .str.lower()
        .str.strip()
        .str.replace(r'[^a-záéíóúñ ]', '', regex=True)
        .str.replace(r'\s+', ' ', regex=True)
)

print(df)
```
