```python
#1. Obtener los dos DataFrames que contienen los datos de ventas.
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd

df_norte = pd.read_excel('/content/drive/MyDrive/datasets/c8_ventas.xlsx', sheet_name='Norte')
df_sur   = pd.read_excel('/content/drive/MyDrive/datasets/c8_ventas.xlsx', sheet_name='Sur')

#df_norte.head()
#df_sur.head()
#print(df_norte)


#2. Concatenar los DataFrames (decidir si correponde hacerlo horizontal o verticalmente), verificar y
#describir el resultado. Sugerir cómo manejar los datos faltantes, si los hubiera.

print("Se decide concatenar los df de manera vertical para que mantengan la misma estructura . En caso de faltantes se eliminarian")
#estandarizo los df:
df_sur.rename(columns={
    'Sede': 'Sucursal',
    'Ventas': 'Ingresos'
}, inplace=True)

#concateno
df_concatenado = pd.concat([df_norte, df_sur], axis = 0)

#chequeo si esta ok
#print(df_concatenado)

#3. segurarse de realizar las operaciones previas necesarias para obtener la siguiente estructura de columnas: Sucursal, Producto, Ventas, Mes
df_concatenado.rename(columns={'Ingresos' : 'Ventas'}, inplace=True)
print(df_concatenado)
```

####Análisis de Inversiones en Gestión de Personal

- 1. Obtener los DataFrames.

- 2. Utiliza pd.merge() para combinar ambos DataFrames basados en la columna que se considere más apropiada, con el join que corresponda. Dejar documentada la justificación de ambas elecciones.

#####Nota al pie: Para no perder los archivos voy a trabajar con una copia que hice y guarde en /datasets


```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd

#1. Obtener los DF:
df_data_empleados = pd.read_csv('/content/drive/MyDrive/datasets/c8_data_empleados.csv')
df_situacion_empleados   = pd.read_csv('/content/drive/MyDrive/datasets/c8_situacion_empleados.csv')


#2. Utiliza pd.merge() para combinar ambos DataFrames basados en la columna que se considere más apropiada,
#con el join que corresponda. Dejar documentada la justificación de ambas elecciones.
df_empleados_merged = pd.merge(df_data_empleados, df_situacion_empleados, on = 'ID', how='inner')
print(df_empleados_merged)
print("Se utiliza ID para mergear dado que esta en todos los campos, además se usa el inner join xq necesito los valores que tengan correspondencia en ambos df")
```
