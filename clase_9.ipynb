#### Actividad 1: Análisis de Tendencia Central

Realizá los siguientes cálculos en Python con los datos sobre las ventas mensuales en unidades del último año:

- 1. Calcular la media de las ventas.

- 2. Calcular la mediana de las ventas.

- 3. Calcular la moda de las ventas.

Interpretá y explicá que significan estos valores.


```python
#Obtener los dos DataFrames
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd

df_norte = pd.read_excel('/content/drive/MyDrive/datasets/c8_ventas.xlsx', sheet_name='Norte')
df_sur   = pd.read_excel('/content/drive/MyDrive/datasets/c8_ventas.xlsx', sheet_name='Sur')


#estandarizo los df:
df_sur.rename(columns={
    'Sede': 'Sucursal',
    'Ventas': 'Ingresos'
}, inplace=True)

#concateno
df_concatenado = pd.concat([df_norte, df_sur], axis = 0)
#renombro
df_concatenado.rename(columns={'Ingresos' : 'Ventas'}, inplace=True)

#borro los $
df_concatenado['Ventas'] = (
    df_concatenado['Ventas']
    .replace(r'[\$,]', '', regex=True)
    .astype(float)
)



# Agrupo por mes
ventas_mensuales = df_concatenado.groupby("Mes")["Ventas"].sum()

# Métricas
#1. media:
media = round(ventas_mensuales.mean(),2)
#2. mediana:
mediana = round(ventas_mensuales.median(),2)
#3. moda:
moda = round(ventas_mensuales.mode(),2)

print("Media:", media)
print("Mediana:", mediana)
print("Moda:", moda)
```

#### Actividad 2: Medidas de Dispersión

Utiliza el mismo conjunto de datos de ventas mensuales del ejercicio anterior y realiza lo siguiente en Python:

- 1. Calcular el rango de las ventas.

- 2. Calcular la varianza de las ventas.

- 3. Calcular la desviación standard de las ventas.

```python
#Obtener los dos DataFrames
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd

df_norte = pd.read_excel('/content/drive/MyDrive/datasets/c8_ventas.xlsx', sheet_name='Norte')
df_sur   = pd.read_excel('/content/drive/MyDrive/datasets/c8_ventas.xlsx', sheet_name='Sur')


#estandarizo los df:
df_sur.rename(columns={
    'Sede': 'Sucursal',
    'Ventas': 'Ingresos'
}, inplace=True)

#concateno
df_concatenado = pd.concat([df_norte, df_sur], axis = 0)
#renombro
df_concatenado.rename(columns={'Ingresos' : 'Ventas'}, inplace=True)

#borro los $
df_concatenado['Ventas'] = (
    df_concatenado['Ventas']
    .replace(r'[\$,]', '', regex=True)
    .astype(float)
)


# Agrupo por mes
ventas_mensuales = df_concatenado.groupby("Mes")["Ventas"].sum()

# Métricas de dispersión
#1. rango:
rango = ventas_mensuales.max() - ventas_mensuales.min()
#2. varianza:
varianza = ventas_mensuales.var()
#3. desvio std:
desv_std = ventas_mensuales.std()

print("Rango:", rango)
print("Varianza:", varianza)
print("Desviación estándar:", desv_std)
```
